```python
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
#if label is 1 it means stock price will increase with respect to particular headlines,if 0 it means it will either constant or decrease
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
data=train.iloc[:,2:27]
```

```python
data.columns
```

```python
# Removing punctuations
#apart from a-z ,remove all by blank
data.replace("[^a-zA-Z]"," ", inplace=True)
```

```python
data.head()
```

```python
# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
new_Index
```

```python
data.columns= new_Index
data.head(5)
```

```python
data.index
```

```python
len(data)
```

```python
### accessing data of 1st column
data['0']
```

```python
# Convertng headlines to lower case
#why str.lower on 0,1,2---- bcz we can apply function on a column only
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
##print entire data of 2nd row
data.iloc[1,0:25]
```

```python
#create a list of all headlines for row 2,as 25 headlines in row 2
' '.join([str(x) for x in data.iloc[1,0:25]])
```

```python
## create a list headlines to insert all the headlines of all rows
headlines=[]
for row in range(0,len(data)):
    headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

```python
#Implementing Bag of Code
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF Code
cv=CountVectorizer(ngram_range=(2,2))
traindata_x=cv.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindata_x,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_data = cv.transform(test_transform)
```

```python
randomclassifier.fit(traindata_x,train['Label'])
```

```python
predictions = randomclassifier.predict(test_data)
predictions
```

```python
#plot confusion matrix
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues, labels=["positive", "negative"]):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45)
    plt.yticks(tick_marks, labels)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plt.figure()
cm=confusion_matrix(test['Label'],predictions)
print(cm)
plot_confusion_matrix(cm)
plt.show()
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

```python
# Sample headlines for testing
sample_headlines = [
    "Economy shows signs of recovery, stocks expected to rise",
    "Tech giant announces record-breaking profits",
]

# Preprocess the sample headlines
sample_headlines_processed = []
for headline in sample_headlines:
    # Convert to lowercase
    headline_lower = headline.lower()
    # Replace non-alphabetic characters with space
    headline_cleaned = ''.join(char if char.isalpha() else ' ' for char in headline_lower)
    # Append to the list of processed headlines
    sample_headlines_processed.append(headline_cleaned)

# Transform the processed headlines using the CountVectorizer
sample_headlines_transformed = cv.transform(sample_headlines_processed)

# Predict the labels for the sample headlines
predictions_sample = randomclassifier.predict(sample_headlines_transformed)

# Print the predictions
for headline, prediction in zip(sample_headlines, predictions_sample):
    if prediction == 1:
        print(f"{headline}: Stock price is predicted to increase")
    else:
        print(f"{headline}: Stock price is predicted to either remain constant or decrease")
```
